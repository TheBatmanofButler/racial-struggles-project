```
RACIAL_DATA_1900_URL = "https://github.com/TheBatmanofButler/racial-struggles-project/raw/master/data/nhgis0005_csv/nhgis0005_ds31_1900_state.csv"
```

```
racial_data_1900 = pd.read_csv(RACIAL_DATA_1900_URL)
```

```
racial_data_1900.head()
```

```
racial_data_1900 = racial_data_1900.drop(columns=["GISJOIN", "STATEA", "COUNTYA", "AREANAME"])
columns = racial_data_1900.columns
print(columns)
```

1900 data is split into males and females for "negro" and "other colored"

```
racial_data_1900 = racial_data_1900.rename(columns={
    "AZ2001": "White",
    "AZ3001": "Other Colored >> Male",
    "AZ3002": "Other Colored >> Female",
    "AZ3003": "Negro >> Male",
    "AZ3004": "Negro >> Female"
})
```

```
racial_data_1900.head()
```

```
racial_data_1900["Other Colored"] = racial_data_1900["Other Colored >> Male"] + racial_data_1900["Other Colored >> Female"]
racial_data_1900["Negro"] = racial_data_1900["Negro >> Male"] + racial_data_1900["Negro >> Female"]
racial_data_1900 = racial_data_1900.drop(columns=["Other Colored >> Male", "Other Colored >> Female", "Negro >> Male", "Negro >> Female"])
```

```
racial_data_1900.head()
```

```
racial_data_territories_1900 = racial_data_1900[racial_data_1900["STATE"].str.contains("Territory")]
```

```
racial_data_territories_1900
```
